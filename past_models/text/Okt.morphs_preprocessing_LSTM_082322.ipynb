```python
voice = pd.read_csv('/content/drive/MyDrive/보이스피싱_검출_AI_프로젝트/데이터 전처리/취합본/preprocessed_done_0816.csv')
```

```python
# nan value 제거
voice = voice.dropna()
```

```python
pre_X_train, pre_X_test, y_train, y_test = train_test_split(voice['text'], voice['label'], test_size=0.2, shuffle=True, random_state= 23)
```

```python
X_train = []
for i,text in enumerate(pre_X_train):
  X_train.append(text.split(","))
  print(i+1,"/",len(pre_X_train),"완료")

X_test = []
for i,text in enumerate(pre_X_test):
  X_test.append(text.split(","))
  print(i+1,"/",len(pre_X_test),"완료")
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
```

```python
print(len(tokenizer.word_counts))
```

```python
tokenizer = Tokenizer(44000)  
```

```python
tokenizer.fit_on_texts(X_train)
```

```python
X_train = tokenizer.texts_to_sequences(X_train)
X_test =  tokenizer.texts_to_sequences(X_test)

#데이터에 대해 정수 인코딩을 진행함 
#X_train에 대해서만 fitting을 진행했기 때문에 X_test도 그 기준으로 인덱싱 됨 



# X_train과 X_test로 나누기 전에 전체 X에 대해 fitting을 시켜줘도 프로세스 상 큰 문제는 없다.



print("# int_encoding done")
```

```python
max_len = 450
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
```

```python
# 모델 구축
#레이어들을 쌓을 모델 생성

model = Sequential()

# 지금까지 구축한 데이터가 Embedding에 들어가 워드 임베딩화 됨 / 4000개의 단어를 32차원으로 보낸다
model.add(Embedding(44000,32))

# 임베딩회된 데이터가 LSTM을 통하서 특정값을 도출함 / 입력은 32차원으로 보내온 임베딩 데이터
model.add(LSTM(32))

# sigmoid함수에 통과시켜서 최종값을 얻음
model.add(Dense(1, activation='sigmoid'))
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose =1 , patience = 5)
```

```python
model_check = ModelCheckpoint('the_best.h5', moniter = 'val_acc', mode= 'max', verbose=1, save_best_only = True)
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['acc'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=1, batch_size=64, callbacks = [early_stop, model_check])
```

```python
print(model.evaluate(X_test, y_test))
```

```python
def regex_check(input):
    import re
    # 정규표현식을 써서 전처리
    input = re.sub(r'\(.*?\)/', " ", input)
    input = re.sub(r"[$?&+:;~=@#|'<>^*()%!/(a-z)*(A-Z)*]", "", input)
    input = input.replace('\\', '')
    input = input.replace('[', '')
    input = input.replace(']', '')
    input = re.sub(r'\s+', " ", input)
    input = re.sub(r"^\s+", "", input)
    input = re.sub(r'\s+$', "", input)
    # 개행문자 제거
    input = re.sub('\n', "", input)
    return input
```

```python
def spell_check(input):
    from hanspell import spell_checker
    input_convert = input.replace('.', '.#').split('#') #문장 단위로 분리
    input_list = [""]
    for i in input_convert:
        if (len(input_list[-1]) + len(i)) < 500:
            input_list[-1] += i
        else:
            input_list.append(i)
    result = spell_checker.check(input_list)
    result = ''
    for j, k in enumerate(input_list):
        a = spell_checker.check([input_list[j]])
        a = a[0].checked
        result = result + a
    return result
```
